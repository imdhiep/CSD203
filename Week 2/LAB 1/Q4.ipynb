```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Q1.1
    def addToHead(self, x):
        new_node = Node(x)
        if not self.head:
            new_node.next = new_node  # Circular link to itself
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next != self.head:
                current_node = current_node.next
            current_node.next = new_node
            new_node.next = self.head
            self.head = new_node

    # Q1.2
    def addToTail(self, x):
        new_node = Node(x)
        if not self.head:
            new_node.next = new_node  # Circular link to itself
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next != self.head:
                current_node = current_node.next
            current_node.next = new_node
            new_node.next = self.head

    # Q1.3
    def addAfter(self, p, x):
        new_node = Node(x)
        if not self.head:
            print("The given previous node must exist in CircularLinkedList.")
            return
        new_node.next = p.next
        p.next = new_node

    # Q1.4
    def traverse(self):
        if not self.head:
            return
        current_node = self.head
        while True:
            print(current_node.data)
            current_node = current_node.next
            if current_node == self.head:
                break

    # Q1.5
    def deleteFromHead(self):
        if not self.head:
            return None
        temp = self.head
        if self.head.next == self.head:
            self.head = None
        else:
            current_node = self.head
            while current_node.next != self.head:
                current_node = current_node.next
            current_node.next = self.head.next
            self.head = self.head.next
        return temp.data

    # Q1.6
    def deleteFromTail(self):
        if not self.head:
            return None
        temp = self.head
        if self.head.next == self.head:
            self.head = None
        else:
            current_node = self.head
            while current_node.next.next != self.head:
                current_node = current_node.next
            current_node.next = self.head
        return temp.data

    # Q1.7
    def deleteAfter(self, p):
        if not self.head or p.next == self.head:
            return None
        temp = p.next
        p.next = temp.next
        return temp.data

    # Q1.8
    def deleteNode(self, x):
        if not self.head:
            return None
        if self.head.data == x:
            return self.deleteFromHead()
        current_node = self.head
        while current_node.next != self.head and current_node.next.data != x:
            current_node = current_node.next
        if current_node.next == self.head:
            return None
        temp = current_node.next
        current_node.next = temp.next
        return temp.data

    # Q1.9
    def search(self, x):
        if not self.head:
            return None
        current_node = self.head
        while True:
            if current_node.data == x:
                return current_node
            current_node = current_node.next
            if current_node == self.head:
                break
        return None

    # Q1.10
    def count(self):
        count = 0
        if not self.head:
            return count
        count = 1
        current_node = self.head.next
        while current_node != self.head:
            count += 1
            current_node = current_node.next
        return count

    # Q1.11
    def deleteAtIndex(self, i):
        if not self.head:
            return None
        if i == 0:
            return self.deleteFromHead()
        count = 0
        current_node = self.head
        while count < i - 1 and current_node.next != self.head:
            current_node = current_node.next
            count += 1
        if current_node.next == self.head:
            return None
        temp = current_node.next
        current_node.next = temp.next
        return temp.data

    # Q1.12
    def sort(self):
        if not self.head:
            return
        current = self.head
        index = Node(None)
        while True:
            index = current.next
            while index != self.head:
                if current.data > index.data:
                    current.data, index.data = index.data, current.data
                index = index.next
            current = current.next
            if current == self.head:
                break

    # Q1.13
    def delNode(self, p):
        if not self.head:
            return None
        if p == self.head:
            return self.deleteFromHead()
        current_node = self.head
        while current_node.next != self.head and current_node.next != p:
            current_node = current_node.next
        if current_node.next == self.head:
            return None
        temp = current_node.next
        current_node.next = temp.next
        return temp.data

    # Q1.14
    def toArray(self):
        arr = []
        current_node = self.head
        if not self.head:
            return arr
        while True:
            arr.append(current_node.data)
            current_node = current_node.next
            if current_node == self.head:
                break
        return arr

    # Q1.15
    def mergeLists(self, other):
        if not self.head:
            return other.head
        if not other.head:
            return self.head
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = other.head
        temp = other.head
        while temp.next != other.head:
            temp = temp.next
        temp.next = self.head
        return self.head

    # Q1.16
    def addBefore(self, p, x):
        new_node = Node(x)
        if p == self.head:
            new_node.next = self.head
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next != self.head and current_node.next != p:
                current_node = current_node.next
            if current_node.next == self.head:
                return
            new_node.next = p
            current_node.next = new_node

    # Q1.17
    def append(self, other):
        if not self.head:
            self.head = other.head
        elif not other.head:
            return
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = other.head
            temp = other.head
            while temp.next != other.head:
                temp = temp.next
            temp.next = self.head

    # Q1.18
    def max(self):
        if not self.head:
            return None
        max_value = self.head.data
        current_node = self.head.next
        while current_node != self.head:
            if current_node.data > max_value:
                max_value = current_node.data
            current_node = current_node.next
        return max_value

    # Q1.19
    def min(self):
        if not self.head:
            return None
        min_value = self.head.data
        current_node = self.head.next
        while current_node != self.head:
            if current_node.data < min_value:
                min_value = current_node.data
            current_node = current_node.next
        return min_value

    # Q1.20
    def sum(self):
        total = 0
        if not self.head:
            return total
        total += self.head.data
        current_node = self.head.next
        while current_node != self.head:
            total += current_node.data
            current_node = current_node.next
        return total

    # Q1.21
    def avg(self):
        total = self.sum()
        count = self.count()
        if count != 0:
            return total / count
        else:
            return None

    # Q1.22
    def sorted(self):
        if not self.head:
            return True
        current_node = self.head
        while current_node.next != self.head:
            if current_node.data > current_node.next.data:
                return False
            current_node = current_node.next
        return True

    # Q1.23
    def insert(self, x):
        new_node = Node(x)
        if not self.head:
            new_node.next = new_node  # Circular link to itself
            self.head = new_node
        elif x <= self.head.data:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            new_node.next = self.head
            temp.next = new_node
            self.head = new_node
        else:
            current_node = self.head
            while current_node.next != self.head and current_node.next.data < x:
                current_node = current_node.next
            new_node.next = current_node.next
            current_node.next = new_node

    # Q1.24
    def reverse(self):
        if not self.head:
            return
        prev_node = None
        current_node = self.head
        next_node = None
        while True:
            next_node = current_node.next
            current_node.next = prev_node
            prev_node = current_node
            current_node = next_node
            if current_node == self.head:
                break
        self.head = prev_node

    # Q1.25
    def sameContents(self, other):
        current1 = self.head
        current2 = other.head
        while True:
            if (current1 is None and current2 is not None) or (current1 is not None and current2 is None):
                return False
            if current1.data != current2.data:
                return False
            current1 = current1.next
            current2 = current2.next
            if current1 == self.head and current2 == other.head:
                break
        return True
```
