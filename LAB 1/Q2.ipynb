```python
class Node:
   def __init__(self, value, next_node=None):
       self.value = value
       self.next = next_node
```

```python
class SinglyLinkedList:
   def __init__(self):
       self.head = None
    #Q2.1
   def addToHead(self, x):
       new_node = Node(x, self.head)
       self.head = new_node
    #Q2.2
   def addToTail(self, x):
       new_node = Node(x)
       if self.head is None:
           self.head = new_node
       else:
           current = self.head
           while current.next:
               current = current.next
           current.next = new_node
    #Q2.3
   def addAfter(self, p, x):
       new_node = Node(x)
       current = self.head
       while current and current.value != p:
           current = current.next
       if current:
           new_node.next = current.next
           current.next = new_node
    #Q2.4
   def traverse(self):
       current = self.head
       while current:
           print(current.value)
           current = current.next
    #Q2.5
   def deleteFromHead(self):
       if self.head:
           deleted_node = self.head
           self.head = self.head.next
           return deleted_node.value
    #Q2.6
   def deleteFromTail(self):
       if self.head:
           current = self.head
           while current.next:
               if not current.next.next:
                  deleted_node = current.next
                  current.next = None
                  return deleted_node.value
               current = current.next
    #Q2.7
   def deleteAfter(self, p):
       current = self.head
       while current and current.value != p:
           current = current.next
       if current and current.next:
           deleted_node = current.next
           current.next = current.next.next
           return deleted_node.value
    #Q2.8
   def delete(self, x):
       if self.head and self.head.value == x:
           return self.deleteFromHead()
       current = self.head
       while current and current.next and current.next.value != x:
           current = current.next
       if current and current.next:
           deleted_node = current.next
           current.next = current.next.next
           return deleted_node.value
    #Q2.9
   def search(self, x):
       current = self.head
       while current and current.value != x:
           current = current.next
       return current
    #Q2.10
   def count(self):
       count = 0
       current = self.head
       while current:
           count += 1
           current = current.next
       return count
```
