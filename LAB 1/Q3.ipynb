```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    # Q1.1
    def addToHead(self, x):
        new_node = Node(x)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    # Q1.2
    def addToTail(self, x):
        new_node = Node(x)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    # Q1.3
    def addAfter(self, p, x):
        new_node = Node(x)
        if p is None:
            print("The given previous node must exist in LinkedList.")
            return
        new_node.next = p.next
        if p.next:
            p.next.prev = new_node
        p.next = new_node
        new_node.prev = p

    # Q1.4
    def traverse(self):
        current_node = self.head
        while current_node:
            print(current_node.data)
            current_node = current_node.next

    # Q1.5
    def deleteFromHead(self):
        if self.head is None:
            return None
        temp = self.head
        self.head = self.head.next
        if self.head:
            self.head.prev = None
        else:
            self.tail = None
        return temp.data

    # Q1.6
    def deleteFromTail(self):
        if self.head is None:
            return None
        temp = self.tail
        self.tail = self.tail.prev
        if self.tail:
            self.tail.next = None
        else:
            self.head = None
        return temp.data

    # Q1.7
    def deleteAfter(self, p):
        if p is None or p.next is None:
            return None
        temp = p.next
        p.next = temp.next
        if temp.next:
            temp.next.prev = p
        return temp.data

    # Q1.8
    def deleteNode(self, x):
        current = self.head
        while current:
            if current.data == x:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next

                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev

                return current.data
            current = current.next
        return None

    # Q1.9
    def search(self, x):
        current_node = self.head
        while current_node:
            if current_node.data == x:
                return current_node
            current_node = current_node.next
        return None

    # Q1.10
    def count(self):
        count = 0
        current_node = self.head
        while current_node:
            count += 1
            current_node = current_node.next
        return count

    # Q1.11
    def deleteAtIndex(self, i):
        current_node = self.head
        position = 0
        if position == i:
            self.head = current_node.next
            if self.head:
                self.head.prev = None
            else:
                self.tail = None
            return current_node.data
        while current_node and position + 1 != i:
            position += 1
            current_node = current_node.next
        if current_node is None:
            return None
        temp = current_node.next
        current_node.next = temp.next
        if temp.next:
            temp.next.prev = current_node
        return temp.data

    # Q1.12
    def sort(self):
        current = self.head
        index = Node(None)
        if self.head is None:
            return
        else:
            while current is not None:
                index = current.next
                while index is not None:
                    if current.data > index.data:
                        current.data, index.data = index.data, current.data
                    index = index.next
                current = current.next

    # Q1.13
    def deleteNode(self, p):
        if p is None:
            return None
        if p.prev:
            p.prev.next = p.next
        else:
            self.head = p.next

        if p.next:
            p.next.prev = p.prev
        else:
            self.tail = p.prev

        return p.data

    # Q1.14
    def toArray(self):
        arr = []
        current_node = self.head
        while current_node:
            arr.append(current_node.data)
            current_node = current_node.next
        return arr

    # Q1.15
    def mergeLists(self, other):
        if not self.head:
            return other.head
        if not other.head:
            return self.head

        self.tail.next = other.head
        other.head.prev = self.tail
        self.tail = other.tail

        return self.head

    # Q1.16
    def addBefore(self, p, x):
        new_node = Node(x)
        if p is self.head:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        else:
            new_node.prev = p.prev
            new_node.next = p
            p.prev.next = new_node
            p.prev = new_node

    # Q1.17
    def append(self, other):
        if not self.head:
            self.head = other.head
            self.tail = other.tail
        elif other.head:
            self.tail.next = other.head
            other.head.prev = self.tail
            self.tail = other.tail

    # Q1.18
    def max(self):
        max_value = float('-inf')
        current_node = self.head
        while current_node:
            if current_node.data > max_value:
                max_value = current_node.data
            current_node = current_node.next
        return max_value

    # Q1.19
    def min(self):
        min_value = float('inf')
        current_node = self.head
        while current_node:
            if current_node.data < min_value:
                min_value = current_node.data
            current_node = current_node.next
        return min_value

    # Q1.20
    def sum(self):
        total = 0
        current_node = self.head
        while current_node:
            total += current_node.data
            current_node = current_node.next
        return total

    # Q1.21
    def avg(self):
        total = 0
        count = 0
        current_node = self.head
        while current_node:
            total += current_node.data
            count += 1
            current_node = current_node.next
        if count != 0:
            return total / count
        else:
            return None

    # Q1.22
    def sorted(self):
        current_node = self.head
        while current_node and current_node.next:
            if current_node.data > current_node.next.data:
                return False
            current_node = current_node.next
        return True

    # Q1.23
    def insert(self, x):
        new_node = Node(x)
        current_node = self.head
        if current_node is None or current_node.data >= x:
            new_node.next = current_node
            if current_node:
                current_node.prev = new_node
            self.head = new_node
        else:
            while current_node.next is not None and current_node.next.data < x:
                current_node = current_node.next
            new_node.next = current_node.next
            if current_node.next:
                current_node.next.prev = new_node
            current_node.next = new_node
            new_node.prev = current_node

    # Q1.24
    def reverse(self):
        current_node = self.head
        while current_node:
            temp = current_node.next
            current_node.next = current_node.prev
            current_node.prev = temp
            current_node = temp
        temp = self.head
        self.head = self.tail
        self.tail = temp

    # Q1.25
    def sameContents(self, other):
        current1 = self.head
        current2 = other.head
        while current1 and current2:
            if current1.data != current2.data:
                return False
            current1 = current1.next
            current2 = current2.next
        return current1 is None and current2 is None
```
